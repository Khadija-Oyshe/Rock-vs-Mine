```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x=df.drop(60,axis=1)
y=df[60]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
pipe_logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=10000))
])
pipe_svc = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVC())
])
pipe_knn = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsClassifier())
])
pipe_rf = Pipeline([
    ('model', RandomForestClassifier(random_state=42))
])
```

```python
model_hyperparameters = {

    pipe_logreg: {
        'model__C': [0.1, 1, 10]
    },

    pipe_svc: {
        'model__C': [0.1, 1, 10],
        'model__kernel': ['linear', 'rbf']
    },

    pipe_knn: {
        'model__n_neighbors': [3, 5, 7]
    },

    pipe_rf: {
        'model__n_estimators': [50, 100],
        'model__max_depth': [None, 10, 20]
    }
}
```

```python
def ModelSelection(model_param_dict, X_train, y_train):

    results = []

    for model, params in model_param_dict.items():

        grid = GridSearchCV(model, params, cv=5)
        grid.fit(X_train, y_train)

        results.append({
            'Model': model.named_steps['model'].__class__.__name__,
            'Best CV Score': grid.best_score_,
            'Best Parameters': grid.best_params_
        })

    return pd.DataFrame(results)
```

```python
results_df = ModelSelection(model_hyperparameters, X_train, y_train)

print(results_df)
```

```python
best_model = grid.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
